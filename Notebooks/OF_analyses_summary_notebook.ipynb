## Aggregate encoding model scores into a single pandas frame

```python
%%time

overwrite = False

def get_model_summmary_scores(subjects = None, overwrite = False, path="/mnt/Data_HD2T/SummaryData/"):

    if subjects is None:
        subjects = subjects = ['Li', 'Ne', 'Cl', 'Mi', 'Al', 'Ca']

    model_scores_fn = Path(path,'of_model_scores_summary_table.csv')
    
    if overwrite:
        models = ['speed', 'hd', 'border', 'grid', 'pos']
        metrics = ['r2', 'map_r', 'n_err']
        splits = ['train', 'test']
        n_models = len(models)
        n_metrics = len(metrics)
        n_splits = len(splits)
        n_vals_per_unit = n_models*n_metrics*n_splits

        unit_count = 0
        model_scores = pd.DataFrame()
        for subject in subjects:
            subject_info = SubjectInfo(subject)
            for session in subject_info.sessions:
                if 'OF' in session:
                    session_info = SubjectSessionInfo(subject, session)
                    n_session_units = session_info.n_units
                    if n_session_units>0:
                        try:
                            temp = session_info.get_encoding_models()
                            mask = temp['metric'].isin(metrics) & temp['model'].isin(models)

                            session_models_scores = pd.DataFrame(index=np.arange(n_session_units*n_vals_per_unit), 
                                                          columns=['unit_id','subject','session','session_unit_id','unit_type', 
                                                                   'tt', 'tt_cl', 'model', 'split', 'metric', 'value'])
                            session_models_scores.loc[:,['model', 'split', 'metric', 'value']] = \
                            temp.loc[mask, ['model', 'split', 'metric', 'value']].values

                            session_models_scores['session'] = session_info.session
                            session_models_scores['subject'] = session_info.subject
                            session_models_scores['session_unit_id'] = temp.loc[mask, 'unit_id'].values

                            session_models_scores['unit_id'] = np.tile(np.arange(n_session_units),n_vals_per_unit)+unit_count

                            session_models_scores['unit_type'] = [v[0] for k, v in session_info.cluster_ids.items()]*n_vals_per_unit
                            session_models_scores['tt'] = [v[1] for k, v in session_info.cluster_ids.items()]*n_vals_per_unit
                            session_models_scores['tt_cl'] = [v[2] for k, v in session_info.cluster_ids.items()]*n_vals_per_unit

                            model_scores = model_scores.append(session_models_scores)
                            unit_count += n_session_units
                        except ValueError:
                            pass
                        finally:
                            pass
        model_scores=model_scores.reset_index(drop=True)
        model_scores=model_scores.astype({"value": float})
        model_scores.to_csv(model_scores_fn)

    else:
        model_scores = pd.read_csv(model_scores_fn, index_col=0)

    return model_scores

models_scores = get_model_summmary_scores()
```

```python
def match_unit_ids(metric_scores, models_scores, path="/mnt/Data_HD2T/SummaryData/"):
    session_unit_id_array = metric_scores[['session','session_unit_id']].values
    session_unit_id_tuple= [tuple(ii) for ii in session_unit_id_array]
    sid_2_uid = {}
    uid_2_sid = {}

    used_ids = []
    unique_id_cnt = 0
    for suid in session_unit_id_tuple:
        if not suid in used_ids:
            sid_2_uid[suid] = unique_id_cnt
            uid_2_sid[unique_id_cnt] = suid

            unique_id_cnt += 1
            used_ids += [suid]
           
    session_unit_id_array = models_scores[['session','session_unit_id']].values
    session_unit_id_tuple= [tuple(ii) for ii in session_unit_id_array]
    models_scores['unit_id']=[sid_2_uid[suid] for suid in session_unit_id_tuple]

    model_scores_fn = Path(path,'of_model_scores_summary_table.csv')
    models_scores.to_csv(model_scores_fn)
    
    return models_scores, sid_2_uid, uid_2_sid

models_scores, sid_2_uid, uid_2_sid = match_unit_ids(metric_scores, models_scores)
```

```python
def unit_counts_table(unit_type, stab_corr=0.5):
    of_metrics = ['speed', 'hd', 'border', 'grid', 'stability']
    unit_count_table = pd.DataFrame(index=subjects, columns=of_metrics)
    for subject in subjects:
        for analysis in of_metrics:
            idx = (metric_scores['analysis_type']==analysis) & (metric_scores['subject']==subject) & (metric_scores['unit_type']==unit_type)
            if analysis=="stability":
                unit_count_table.at[subject, analysis] = (metric_scores[idx]['score']>=stab_corr).sum()
            else:
                unit_count_table.at[subject, analysis] =  metric_scores[idx]['sig'].sum()

    return unit_count_table
    
```

```python
@interact(x_metric=metrics, y_metric=metrics,  x_split=splits, y_split=splits[::-1], x_model=models, y_model=models, unit_type=['all', 'cell', 'mua'])
#@interact(metric=metrics, split=splits, models=['all']+models, unit_type=['all', 'cell', 'mua'])
def plot_model_score_metrics(x_metric, y_metric,  x_split, y_split,  x_model='pos', y_model='pos', unit_type='cell', by_subject=False):
    if unit_type == 'all':
        x_subset = models_scores['metric'].isin([x_metric]) & models_scores['split'].isin([x_split]) & models_scores['model'].isin([x_model])
        y_subset = models_scores['metric'].isin([y_metric]) & models_scores['split'].isin([y_split]) & models_scores['model'].isin([y_model])
    else:
        x_subset = models_scores['metric'].isin([x_metric]) & models_scores['split'].isin([x_split]) & models_scores['model'].isin([x_model]) & models_scores['unit_type'].isin([unit_type])
        y_subset = models_scores['metric'].isin([y_metric]) & models_scores['split'].isin([y_split]) & models_scores['model'].isin([y_model]) & models_scores['unit_type'].isin([unit_type])

    data = models_scores.loc[x_subset, ['subject', 'unit_type', 'model', 'value']].copy()
    data.loc[:, 'x_var'] = data['value'].values.astype(float)
    data.loc[:, 'y_var'] = models_scores.loc[y_subset, 'value'].values.astype(float)
    data.dropna(inplace=True)
    
    max_val = np.max( np.array( (data.loc[:,'x_var'].max(), data.loc[:,'y_var'].max())) )

    if by_subject:
        g = sns.JointGrid(data=data, x="x_var", y="y_var", hue='subject')
        
        g.plot_joint(sns.scatterplot,alpha=0.75)
        g.plot_marginals(sns.kdeplot, **{'bw_adjust':0.5})
        g.plot_marginals(sns.rugplot, **{'height':0.15, 'alpha':0.1})
    else:
        g = sns.JointGrid(data=data, x="x_var", y="y_var")
        g.plot_joint(sns.regplot, **{'scatter_kws': {'alpha':0.2}})
        g.plot_marginals(sns.kdeplot, **{'bw_adjust':0.5})
        g.plot_marginals(sns.rugplot, **{'height':0.15, 'alpha':0.1, 'color':'grey'})
        
    if x_metric==y_metric:
        
        g.ax_joint.plot([0, max_val*1.05],[0, max_val*1.05], '--', linewidth=2, alpha=0.8, color = 'k')
        
    g.ax_joint.set(
                    xlabel = f"{x_metric} {x_model} {x_split} ",
                    ylabel = f"{y_metric} {y_model} {y_split}",
                    )
```

```python
@interact(metric_type=analyses,model=models, model_metric=metrics, split=splits, hue=['None', 'subject', 'unit_type'] ,unit_type=['all', 'cell', 'mua'])
def plot_model_vs_metrics(metric_type, model, model_metric, split, unit_type='cell', hue='None', linreg=False):
    
    if unit_type == 'all':
        x_subset = (models_scores['model']==model) & (models_scores['metric']==model_metric) & \
            (models_scores['split']==split)
        unit_ids = models_scores[model_subset].unit_id.unique()
        y_subset = (metric_scores['analysis_type']==metric_type) & (metric_scores['unit_id'].isin(unit_ids))
    else:
        x_subset = (models_scores['model']==model) & (models_scores['unit_type']==unit_type) & \
            (models_scores['metric']==model_metric) & (models_scores['split']==split)
        unit_ids = models_scores[model_subset].unit_id.unique()
        y_subset = (metric_scores['analysis_type']==metric_type) & (metric_scores['unit_type']==unit_type) & \
            (metric_scores['unit_id'].isin(unit_ids))

    data = models_scores.loc[x_subset, ['subject', 'unit_type', 'model', 'value']].copy()
    data.loc[:, 'x_var'] = data['value'].values.astype(float)
    data.loc[:, 'y_var'] = metric_scores.loc[y_subset, 'score'].values.astype(float)
    data.dropna(inplace=True)

    f, ax = plt.subplots(figsize=(4,4))
    if hue=='None':
        if linreg:
            sns.regplot(data=data,x='x_var', y='y_var', color='k', scatter_kws={'alpha':0.2})
        else:
            sns.scatterplot(data=data,x='x_var', y='y_var', color='k', alpha=0.5)
    else:
        if linreg:
            sns.regplot(data=data,x='x_var', y='y_var', color='k', hue=hue, scatter_kws={'alpha':0.2})
        else:
            sns.scatterplot(data=data,x='x_var', y='y_var', hue=hue,alpha=0.5, color='k')

    ax.set_xlabel(f'{model} model {model_metric} ({split}) ')
    ax.set_ylabel(f'{metric_type} score')
    ax.set_title(f'Model vs Score')

    plot_str = f"model-{model}-{model_metric}-{split}_vs_metric-{metric_type}_{unit_type}"
    return f, plot_str
```
